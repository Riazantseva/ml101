```
caps = cv.imread('caps.jpg')
hsv = cv.cvtColor(caps, cv.COLOR_BGR2HSV)

lower_blue = np.array([110,50,50])
upper_blue = np.array([130,255,255])

# Threshold the HSV image to get only blue colors
mask = cv.inRange(hsv, lower_blue, upper_blue)

# Bitwise-AND mask and original image
blue_cap = cv.bitwise_and(caps, caps, mask = mask)
#cv2_imshow(caps)
#cv2_imshow(mask)
#cv2_imshow(blue_cap)



height, width = caps.shape[:2]
caps_scaled1 = cv.resize(caps,(600, 800), interpolation = cv.INTER_CUBIC) 
caps_scaled2= cv.resize(mask,(600, 800), interpolation = cv.INTER_CUBIC)
caps_scaled3= cv.resize(blue_cap,(600, 800), interpolation = cv.INTER_CUBIC)
# only int

cv2_imshow(caps_scaled1)
cv2_imshow(caps_scaled2)
cv2_imshow(caps_scaled3)
```

```
plast = cv.imread("nobg.png")
cv2_imshow(plast)
```

```
low_threshold = 70 # pixel value 
ratio = 3
canny_output = cv.Canny(plast, low_threshold, low_threshold * ratio)
_, contours, _ = cv.findContours(canny_output, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
cv2_imshow(canny_output)
```

```
w = 1200
h = 1200
with open("allContours.svg", "w+") as f:
    f.write(f'<svg width="{w}" height="{h}" xmlns="http://www.w3.org/2000/svg">')

    for c in contours:
        f.write('<path d="M')
        for i in range(len(c)):
            x, y = c[i][0]
            f.write(f"{x} {y} ")
        f.write('" style="stroke:pink"/>')
    f.write("</svg>")
```
